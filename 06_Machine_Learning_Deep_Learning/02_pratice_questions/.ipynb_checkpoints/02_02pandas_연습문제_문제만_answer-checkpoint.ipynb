# 타이타닉 자료에서 그룹별 연산하기

원격 DB에 접속해서 titanic 데이터베이스 안에 있는 p_info, t_info, survived 테이블을 불러와 합치고 아래의 문제를 풀어보세요.<br>
<id:kict00, pw:kict00, ip:221.155.18.165:3306, db:titanic>

```python
import os
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
from dotenv import load_dotenv
load_dotenv()
```

```python
dbid = os.getenv('dbid2')
dbpw = os.getenv('dbpw2')
host = os.getenv('host2')
port = os.getenv('port2')
dbname = os.getenv('dbname')
```

```python
engine = create_engine(f'mysql+pymysql://{dbid}:{dbpw}@{host}:{port}/{dbname}')
conn = engine.connect()
```

```python
table1 = pd.read_sql('p_info', con=conn)
table2 = pd.read_sql('t_info', con=conn)
table3 = pd.read_sql('survived', con=conn)
```

```python
merged_table = pd.merge(table1, table2, how='inner', on='PassengerId')
merged_table
```

```python
df = pd.merge(merged_table, table3, how='inner', on='PassengerId')
df
```
