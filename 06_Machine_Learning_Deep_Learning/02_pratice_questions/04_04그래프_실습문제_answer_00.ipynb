```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('../data/04_graph_practice_dataset.csv', encoding='cp949')
df['날짜'] = pd.to_datetime(df['날짜'])
df.head()
```

## 문제 1️⃣: 월별 전체 매출 추세 (Line Plot)

- `날짜`별 총 `매출`을 선 그래프로 나타내세요.
- `marker='o'`, `linewidth=2`, `grid=True`를 적용하세요.
- 그래프 제목: `월별 매출 추이`

```python
# TODO: 월별 매출 추세 그래프를 그리세요.

# 월별로 그룹화하여 매출 합계 계산
monthly_sales = df.groupby(df['날짜'].dt.to_period('M'))['매출'].sum().reset_index()
monthly_sales['날짜'] = monthly_sales['날짜'].dt.to_timestamp()

x = monthly_sales['날짜']
y = monthly_sales['매출']

fig, ax = plt.subplots(figsize=(5, 3))

# 스타일 지정
ax.plot(x,y,
         color='blue', 
         linewidth=2,
         marker='o',
         label='월별 총 매출')
ax.set_xlabel('날짜')
ax.set_ylabel('매출')
ax.set_title('월별 매출 추이')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

## 문제 2️⃣: 지역별 매출 추세 비교

- 각 `지역`별로 월별 매출 합계를 구하고 한 그래프에 여러 선으로 그리세요.
- 범례(legend)를 추가하세요.
- 제목: `지역별 월별 매출 비교`

```python
# TODO: 지역별 월별 매출을 선 그래프로 여러 개 겹쳐 그리세요.

# (1)지역별, 월별로 그룹화하여 매출 합계 계산
monthly_sales = df.groupby([df['날짜'].dt.to_period('M'), '지역'])['매출'].sum().reset_index()

# (2)날짜를 timestamp 형식으로 전환
monthly_sales['날짜'] = monthly_sales['날짜'].dt.to_timestamp()

# (3)그래프 크기 설정
fig, ax = plt.subplots(figsize=(10, 6))

# (4)각 지역별로 선 그래프 설정
region_colors = {'광주': 'blue','대구':'red','부산':'orange','서울':'green'}


for region in monthly_sales['지역'].unique():
    region_data = monthly_sales[monthly_sales['지역'] == region]
    ax.plot(region_data['날짜'], 
            region_data['매출'],
            linewidth=2,
            marker='o',
            color=region_colors[region],
            label=region)

# (5)라벨 및 제목 설정
ax.set_xlabel('날짜')
ax.set_ylabel('매출')
ax.set_title('지역별 월별 매출 비교')

# (6)주석 설정
ax.legend()

# (7)그리드 설정
ax.grid(True)

# (8)자동 레이아웃 맞춤 설정
plt.tight_layout()
plt.show()
```
