# matplotlib의 2가지 코딩 스타일

## 00-스타일: matplotlib 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot스타일보다 복잡함

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
print(fig)
print(ax)
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(5, 12))
print(fig)
print(ax)
ax[0].plot(x, x, label="linear")
ax[1].plot(x, x**2, label="quadratic")
ax[2].plot(x, x**3, label='cubic')
ax[0].set_xlabel('x label')
ax[0].set_ylabel('y label')
ax[0].set_title('간단 그래프')
ax[0].legend()
ax[1].set_xlabel('x label')
ax[1].set_ylabel('y label')
ax[1].set_title('quadratic 간단 그래프')
ax[1].legend()
ax[2].set_xlabel('x label')
ax[2].set_ylabel('y label')
ax[2].set_title('cubic 간단 그래프')
ax[2].legend()
plt.tight_layout()
plt.show()
```

# 선 모양 바꾸기 set_linestyle(모양)

```python
data1, data2, data3, data4 = np.random.rand(4, 100)
```

* 선 모양: '-', '--', '-.', ':', ''

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle=":")
ax.plot(x, np.cumsum(data2), color='orange', linewidth=2, linestyle="-.")
plt.show()
```

# 산점도 그래프 그리고 색 변경하기
* facecolor(안쪽색), edgecolor(테두리색)

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1, data2, s=50, facecolor="C4", edgecolor='k')
plt.show()
```

* 컬러를 RGB값으로 주고 싶을 때(0, 1, 0.5) 처럼 0-1사이의 숫자를 R, G, B순서의 튜플로 입력

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1, data2, s=50, facecolor=(0,0.5,1), edgecolor=(1,0,0))
plt.show()
```

# X축, Y축 이름, 눈금, 그래프 이름

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
n, bins, patches = ax.hist(x, 50, density=True, facecolor="C0", alpha=0.75)
ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths \n (not really)')
ax.text(70, 0.025, r'$\mu=115, \ \sigma=15$')
ax.axis([60, 160, 0, 0.03])
plt.show()
```

# 눈금자 단위 구간 설정하기

```python
xdata = np.arange(len(data1))
fig, ax = plt.subplots(2, 1, figsize=(5, 2.7), layout='constrained')
ax[0].plot(xdata, data1)
ax[0].set_title('Automatic tics')

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title("Manual ticsks")
plt.show()
```
