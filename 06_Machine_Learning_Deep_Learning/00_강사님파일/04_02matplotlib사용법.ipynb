## 00-스타일: matplotlib 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot 스타일보다 복잡함

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(5, 2.7))
print(ax)
ax[0].plot(x, x, label='linear')
ax[1].plot(x, x**2, label="quadratic")
ax[2].plot(x, x**3, label="cubic")
ax[0].set_xlabel('x label')
ax[0].set_ylabel('y label')
ax[1].set_title('간단 그래프')
ax[1].set_xlabel('x label')
ax[1].set_ylabel('y label')
ax[1].set_title('간단 그래프')
ax[2].set_xlabel('x label')
ax[2].set_ylabel('y label')
ax[2].set_title('간단 그래프')
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

## 0. 준비하기: 기본 세팅

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 보기 좋은 스타일
plt.style.use('seaborn-v0_8')

# 한글 폰트 설정 (윈도우: 맑은 고딕 / mac: AppleGothic 등으로 바꿔도 됨)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

print('Matplotlib version:', plt.matplotlib.__version__)
```

## 1. 라인 플롯 (Line Plot)
### 1-1. 기본 선 그래프

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 9, 14, 15]

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title('기본 선 그래프')
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.show()
```

### 1-2. 스타일 바꾸기 (색, 마커, 선굵기 등)

```python
x = np.arange(1, 11)
y1 = x * 2
y2 = x ** 1.5

plt.figure(figsize=(7,4))
plt.plot(x, y1, color='red', linestyle='--', marker='o', linewidth=2, label='2배 증가')
plt.plot(x, y2, color='blue', linestyle='-', marker='s', linewidth=2, label='제곱근 증가')
plt.title('선 스타일 비교')
plt.xlabel('시간(일)')
plt.ylabel('지표 값')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. 막대 그래프 (Bar Chart)
### 2-1. 기본 세로 막대

```python
categories = ['A제품', 'B제품', 'C제품', 'D제품']
sales = [120, 90, 150, 80]

plt.figure(figsize=(6,4))
plt.bar(categories, sales)
plt.title('제품별 매출 (단위: 만원)')
plt.xlabel('제품')
plt.ylabel('매출')
plt.show()
```

### 2-2. 가로 막대 (`barh`)

```python
plt.figure(figsize=(6,4))
plt.barh(categories, sales)
plt.title('제품별 매출 (가로 막대)')
plt.xlabel('매출(만원)')
plt.ylabel('제품')
plt.show()
```

### 2-3. 그룹 막대 (브랜드 비교)

```python
labels = ['1분기', '2분기', '3분기', '4분기']
brand_a = [100, 120, 90, 140]
brand_b = [80, 110, 95, 130]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(7,4))
plt.bar(x - width/2, brand_a, width=width, label='브랜드 A')
plt.bar(x + width/2, brand_b, width=width, label='브랜드 B')
plt.xticks(x, labels)
plt.title('분기별 매출 비교')
plt.xlabel('분기')
plt.ylabel('매출(만원)')
plt.legend()
plt.show()
```

## 3. 원형 그래프 (Pie Chart)

```python
labels = ['광고', '인건비', '재료비', '기타']
costs = [40, 25, 20, 15]

plt.figure(figsize=(5,5))
plt.pie(costs, labels=labels, autopct='%.1f%%', startangle=90,
        wedgeprops={'edgecolor': 'white'})
plt.title('비용 구성비')
plt.show()
```

## 4. 히스토그램 (값의 분포 보기)

```python
np.random.seed(42)
scores = np.random.normal(loc=75, scale=10, size=200)  # 평균 75점, 표준편차 10

plt.figure(figsize=(7,4))
plt.hist(scores, bins=15, edgecolor='black')
plt.title('시험 점수 분포')
plt.xlabel('점수')
plt.ylabel('학생 수')
plt.show()
```

## 5. 박스플롯 (Boxplot)

```python
class_a = np.random.normal(70, 8, 50)
class_b = np.random.normal(78, 6, 50)
class_c = np.random.normal(74, 10, 50)

plt.figure(figsize=(6,4))
plt.boxplot([class_a, class_b, class_c], labels=['A반', 'B반', 'C반'])
plt.title('반별 시험 점수 분포 비교')
plt.ylabel('점수')
plt.show()
```

## 6. 산점도 & 버블 차트
### 6-1. 산점도 기본

```python
study_hours = np.random.uniform(1, 5, 30)
scores = 50 + study_hours * 10 + np.random.normal(0, 3, 30)

plt.figure(figsize=(6,4))
plt.scatter(study_hours, scores)
plt.title('공부시간과 점수의 관계')
plt.xlabel('공부시간(시간)')
plt.ylabel('시험 점수')
plt.show()
```

### 6-2. 버블 차트 (점 크기에 의미 부여)

```python
size = np.random.uniform(50, 300, 30)  # 버블 크기(예: 과제 제출 수)

plt.figure(figsize=(6,4))
plt.scatter(study_hours, scores, s=size, alpha=0.6, edgecolor='black')
plt.title('공부시간 vs 점수 (버블=과제수)')
plt.xlabel('공부시간(시간)')
plt.ylabel('시험 점수')
plt.show()
```

## 7. Subplot으로 여러 그래프 한 화면에

```python
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].plot(x, y_sin, label='sin')
axes[0].set_title('sin(x)')
axes[0].legend()

axes[1].plot(x, y_cos, color='orange', label='cos')
axes[1].set_title('cos(x)')
axes[1].legend()

plt.suptitle('삼각함수 비교')
plt.show()
```

## 8. 한 그래프에 두 축 사용 (twinx)
매출(억원) vs 방문자수(천 명) 같이 단위 다른 데이터 비교할 때 많이 씀.

```python
months = ['1월','2월','3월','4월','5월','6월']
revenue = [12, 15, 14, 18, 20, 22]   # 억원
visitors = [200, 240, 230, 260, 300, 320]  # 천 명

fig, ax1 = plt.subplots(figsize=(7,4))

ax1.set_title('월별 매출 & 방문자 수')
ax1.set_xlabel('월')
ax1.set_ylabel('매출(억원)', color='tab:blue')
ax1.plot(months, revenue, marker='o', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('방문자 수(천 명)', color='tab:red')
ax2.bar(months, visitors, alpha=0.3, color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.show()
```

## 9. 날짜(시계열) 데이터 시각화

```python
import matplotlib.dates as mdates
from datetime import datetime, timedelta
np.random.seed(0)

dates = [datetime(2025,10,1) + timedelta(days=i) for i in range(14)]
daily_sales = np.cumsum(np.random.randint(5, 20, size=len(dates)))  # 누적 매출 느낌

plt.figure(figsize=(8,4))
plt.plot(dates, daily_sales, marker='o')
plt.title('일별 매출 추이')
plt.xlabel('날짜')
plt.ylabel('매출(누적, 만원)')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=2))
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 13. 이동평균선 (주가 스타일 분석)
5일선 / 10일선 같은 이동평균선은 매출 추세, 유저수 추세 분석에도 그대로 씀.

```python
np.random.seed(1)
days = pd.date_range(start='2025-09-01', periods=30, freq='D')
price = 100 + np.cumsum(np.random.randn(30))  # 랜덤 주가 느낌

stock_df = pd.DataFrame({'날짜': days, '종가': price})
stock_df['MA5'] = stock_df['종가'].rolling(window=5).mean()
stock_df['MA10'] = stock_df['종가'].rolling(window=10).mean()

import matplotlib.dates as mdates

plt.figure(figsize=(9,4))
plt.plot(stock_df['날짜'], stock_df['종가'], label='종가', marker='o', linewidth=1)
plt.plot(stock_df['날짜'], stock_df['MA5'], label='5일 이동평균', linewidth=2)
plt.plot(stock_df['날짜'], stock_df['MA10'], label='10일 이동평균', linewidth=2)

plt.title('주가와 이동평균선 예시')
plt.xlabel('날짜')
plt.ylabel('가격(가상)')
plt.legend()
plt.grid(True)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
