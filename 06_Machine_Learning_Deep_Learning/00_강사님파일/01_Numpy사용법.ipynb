* np.sort()

```python
arr = np.array([
    [[5,2,8],
    [1,6,7]],
    [[4,9,3],
    [0,2,1]]
])
print(arr, arr.shape)
```

```python
# np.sort의 기본 정렬: 가장 안쪽 차원 1차원 배열을 정렬
# axis = -1 (기본값)
print(arr)
np.sort(arr)
```

```python
# axis = 0 2차원 배열에서 같은 행의 요소끼리 정렬
print(arr)
np.sort(arr, axis=0)
```

# argsort()
* 자료중 값을 작은 것에서 큰 순으로 **원래 인덱스** 출력

```python
g = np.array([5, 0.5, -7, 42, 10, 0])
print(g)
```

# np.where(조건, 참, 거짓) 3항 연산 

```python
nums = np.arange(1, 11)
nums
```

```python
result = []
for num in nums:
    if num % 2 == 0:
        result.append("짝수")
    else:
        result.append("홀수")
result
```

```python
result2 = ["짝수" if num % 2 == 0 else "홀수" for num in nums]
result2
```

```python
result3 = np.where(nums % 2 == 0, "짝수", "홀수")
result3
```

# 배열 합치기 np.concatenate()

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])
print(a, a.ndim, a.shape)
print(b, b.ndim, b.shape)
```

* np.concatenate((a, b), axis=0/1) 기본값은 axis=0
* axis=0 아래로 합쳐짐, 행이 늘어남
* axis=1 옆으로 합쳐짐, 열이 늘어남

```python
# axis=0 아래로 늘어나면서 합쳐짐
np.concatenate((a,b))
```

```python
# axis=1 옆으로 늘어나면서 합쳐짐
np.concatenate((a,b), axis=1)
```

# 축(axis) 기반 집계 연산 np.sum()
* axis=0 세로(열) 단위 연산
* axis=1 가로(행) 단위 연산

```python
mat = np.array([[1,2,3], [4,5,6]])
print(mat, mat.ndim, mat.shape)
```

```python
print("전체 요소의 합: ", np.sum(mat))
print("열별 합(axis=0): ", np.sum(mat, axis=0))
print("행별 합(axis=1): ", np.sum(mat, axis=1))
```

# np.mean() 평균

```python
print("전체 요소의 평균: ", np.mean(mat))
print("열별 평균(axis=0): ", np.mean(mat, axis=0))
print("행별 평균(axis=1): ", np.mean(mat, axis=1))
```

# 누적합/ 누적곱 (cumsum/cumprod)
* comsum: 앞에서부터 계속 더한 결과(현금 잔고 추이 등)
* comprod: 앞에서부터 계속 곱한 결과(누적 성장률)

```python
cashflow = np.array([100, -20, 50, -10])
print("현금흐름: ", cashflow)
print("누적잔고: ", np.cumsum(cashflow))
```

```python
# 매년 매출 성장 비율
rates = np.array([10.5, 13, 0.5, 1.3])
print("연별 성장률: ", rates)
print("누적 성장률: ", np.cumprod(rates))
```

# 분산(var), 표준편차(std)

```python
classA = np.array([70, 72, 68, 69, 71])
classB = np.array([50, 90, 30, 100, 10])
```

```python
print("A반 평균: ", np.mean(classA))
print("B반 평균: ", np.mean(classB))
```

```python
print("A반 표준편차: ", np.std(classA))
print("B반 표준편차: ", np.std(classB))
```

```python
print("A반 분산: ", np.var(classA))
print("B반 분산: ", np.var(classB))
```
