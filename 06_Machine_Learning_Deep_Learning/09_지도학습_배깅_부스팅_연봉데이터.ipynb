# ë°ì´í„° ë¶„ì„ ëª¨ë¸ì˜ ì„±ëŠ¥í‰ê°€
* accuracy(ì •í™•ë„/ì •ë¶„ë¥˜ìœ¨): 1ì„ 1ì´ë¼ê³  ë§ì¶”ê³  0ì„ 0ì´ë¼ê³  ë§ì¶˜ ë¹„ìœ¨

* recall(ì¬í˜„ìœ¨/ë¯¼ê°ë„): ì‹¤ì œê°’ì„ ì‹¤ì œê°’ìœ¼ë¡œ ì˜ ì˜ˆì¸¡í•œ ê²ƒ, ì‹¤ì œ Trueì¤‘ ì˜ˆì¸¡ postiveê°€ ë§ì€ ê²ƒ

# confusion matrix(í˜¼ë™í–‰ë ¬)
|              |  ì˜ˆì¸¡ Positive    | ì˜ˆì¸¡ Nagative     |
|:--------------:|:------------------:|:-------------------:|
| **ì‹¤ì œ True** | TP (True Positive) |FN (False Nagative)|
| **ì‹¤ì œ False** | FP (False Positive)|TN (True Negative) |

* TP: ì‹¤ì œ Ture, ì˜ˆì¸¡ Positive(ì‹¤ì œ ì—°ë´‰ >5ë§Œ, ì˜ˆì¸¡ ì—°ë´‰ >5ë§Œ)
* TN: ì‹¤ì œ False, ì˜ˆì¸¡ Negative(ì‹¤ì œ ì—°ë´‰ <=5ë§Œ, ì˜ˆì¸¡ ì—°ë´‰ <=5ë§Œ)
* FP(1í˜• ì˜¤ë¥˜): ì‹¤ì œ True, ì˜ˆì¸¡ Positive(ì‹¤ì œ ì—°ë´‰ <=5ë§Œ, ì˜ˆì¸¡ ì—°ë´‰ >5ë§Œ )
* FN(2í˜• ì˜¤ë¥˜): ì‹¤ì œ True, ì˜ˆì¸¡ Negative(ì‹¤ì œ ì—°ë´‰ >5ë§Œ, ì˜ˆì¸¡ ì—°ë´‰ <=5ë§Œ)

* ì •í™•ë„(accuracy_score): $ TP + TN \over TP+TN+FP+FN $
* ì˜¤ë¶„ë¥˜ìœ¨(FPR): $ FP+FN \over TP+TN+FP+FN $
* ì •ë°€ë„(precision): $ TP \over TP+FP $
* ì¬í˜„ìœ¨/ë¯¼ê°ë„(recall): $ TP \over TP+FN $
* íŠ¹ì´ë„(Specificity): $ TN \over FP+TN $
* F1-score: $  2 \cdot {Precision \cdot Recall \over Precision + Recall} $
