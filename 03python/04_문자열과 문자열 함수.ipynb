## 문자열 찾기: find / index / in
* find(): 문자열이 없으면 -1 출력
* index(): 문자열이 없으면 에러 발생
* in : 문자열이 있으면 True 없으면 False

```python
s = "나는 파이썬이 좋아요."
```

```python
# 파이썬이라는 단어가 시작되는 익덱스 번호
s.find("파이썬")
```

```python
idx = s.find("파이썬")
s[idx:idx+3] # [3:6]과 같다.
```

```python
# 찾는 단어가 없으면 -1
s.find("자바")
```

```python
s.find("이썬")
```

```python
s.index("파이썬")
```

```python
s.index("이썬")
```
