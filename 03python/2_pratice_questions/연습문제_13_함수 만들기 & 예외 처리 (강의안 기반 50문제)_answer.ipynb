## 2번 문제

```python
# 함수
def greet(name):
    print(f"안녕하세요, {name}님!") # f-string을 햇갈리지 말자.
```

```python
# 출력
greet("민수")
```

## 5번 문제

```python
def join_words(a, b):
    return(f"{a} {b}")
```

```python
print(join_words("Hello", "World"))
```

```python
def join_words(a, b):
    return a + ' ' + b
```

```python
print(join_words("Hello", "World"))
```

## 6번 문제

```python
# 함수
def power(x, n=2):
    return x ** n
```

```python
# 답안
print(power(5))
```

```python
# 답안
print(power(2, 3))
```
