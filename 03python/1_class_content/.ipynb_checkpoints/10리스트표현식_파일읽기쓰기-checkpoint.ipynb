# 다중 반복문을 사용한 리스트 표현식
```
#반복문이 2개일 때
[실행할 코드 반복문 1 반복문 2]
#반복문이 3개일 때
[실행할 코드 반복문 1 반복문 2 반복문 3]
```
* 반복문이 3개 이상 중첩되거나 if else 조건이 복잡하게 붙을 경우 일반 문법 사용
* 코드 해석이 너무 어려워짐

* 2중 반복문으로 구구단 출력하기

````python
gugu = []
for i in range(2, 10):
    for j in range(1, 10):
        gugu.append(f"{i} X {j} = {i*j}")
gugu
````

````python
# 리스트표현식으로 2중 반복문 구구단 만들기
gugu2 = [f"{i} X {j} = {i*j}" for i in range(2, 10) for j in range(1, 10)]
gugu2
````

* 3중 반복문으로 시계 만들기

````python
clock = []
for si in range(0,24):
    for bun in range(0, 60):
        for cho in range(0, 60):
            clock.append(f"{si:02d}:{bun:02d}:{cho:02d}")
clock[:10]
````

````python
clock[-10:]
````

````python
# 리스트표현식으로 3중 for문 시계 만들기
clock2 = [f"{si:02d}:{bun:02d}:{cho:02d}" for si in range(0,24) for bun in range(0, 60) for cho in range(0, 60)]
clock2[:10]
````

# 반복문 + if문의 리스트 표현식
* 필터링 
```python
for ... if 조건 (for문이 앞에, if 조건문이 뒤에, else 없음)
```
* if else + for문의 경우
```python
참일 때 if 조건 else 거짓일 때 for ... (조건 표현식이 앞, for문은 뒤에, if else 함께)
```

## 1) 조건을 이용해 필터링 하기 (if만 쓸 때)

* 1-10 중 짝수만 모으기

````python
# 일반 for + if문
evens = []
for num in range(1, 11):
    if num % 2 == 0:
        evens.append(num)
evens
````

````python
# 리스트 표현식, if문으로 필터링
evens2 = [num for num in range(1, 11) if num % 2 == 0]
evens2
````

## 2) 조건에 따라서 값을 바꾸기 (if ~ else)

* 1-10중 짝수면 짝수 홀수면 홀수를 출력

````python
# for + if
labels = []
for num in range(1, 11):
    if num % 2 == 0:
        labels.append("짝수")
    else:
        labels.append("홀수")
labels
````

````python
# 리스트표현식
# 조건표현식 + for문
labels2 = ["짝수" if num % 2 == 0 else "홀수" for num in range(1, 11)]
labels2
````

````python
print(labels2, type(labels2))
````

````python
labels2[3:6]
````

# 제네레이터 표현식
* 리스트표현식과 문법 동일
* [] 아닌 () 로 만듬
* 1회용
* 만들고 나서 반복문으로 내용을 꺼내면 사라짐
* 오직 순차 접근만 가능, 인덱싱, 슬라이싱 불가
* 데이터가 너무 크거나, 한 번만 쓸 데이터인 경우 사용

````python
# 1-10까지 숫자중 짝수인 것만 모은 리스트를 제네레이터로
evens_g = (num for num in range(1, 11) if num % 2 == 0)
evens_g 
````

* 제네레이터로 만든 결과는 next() 혹은 for 반복문으로 내용을 차례대로 꺼낼 수 있음.
* 끝까지 꺼내고 나면 사라짐

````python
g2 = list(evens_g)
g2
````

````python
# next() 제네레이터에서 자료를 하나씩 꺼낼 때 사용
next(evens_g)
````

# 리스트와 제네레이터 크기 비교

````python
# !pip install pympler
````

````python
import sys
from pympler import asizeof
import psutil, os

# 메모리 사용량 확인 함수
def show_memory(msg):
    process = psutil.Process(os.getpid())
    print(f"{msg}: {process.memory_info().rss / 1024**2:.2f} MB")

# 비교 대상 생성
list_var = [x for x in range(1, 1_000_000)]   # 리스트 내포
gen_var  = (x for x in range(1, 1_000_000))   # 제너레이터 표현식

print("==== sys.getsizeof() 결과 ====")
print("리스트:", sys.getsizeof(list_var))
print("제너레이터:", sys.getsizeof(gen_var))

print("\n==== asizeof.asizeof() 결과 (내부 요소 포함) ====")
print("리스트:", asizeof.asizeof(list_var))
print("제너레이터:", asizeof.asizeof(gen_var))

print("\n==== 프로세스 전체 메모리 사용량 비교 ====")
show_memory("리스트 생성 후")
show_memory("제너레이터 생성 후")
````

* data변수 안에는 리스트 안쪽에 67개의 딕셔너리 세트가 저장되어 있음
* 각 딕셔너리에는 ['ForNum', 'NatNum', 'addrCd', 'gungu', 'resNm', 'rnum', 'sido', 'yyyymm'] 키로 관광지 입장객 정보가 저장되어 있음.
* 67개의 딕셔너리에서 하나의 key당 67개의 value를 리스트 형태로 저장
* pandas를 이용해서 dataframe(sql의 table/excel의 sheet)으로 변환
* {'ForNum' : [                ],
   'NatNum' : [                ],}

````python
print(len(data))
````

````python
data[0]
````

````python
print('ForNum', data[66]['ForNum'])
print('NatNum', data[66]['NatNum'])
print('addrCd', data[66]['addrCd'])
print('gungu', data[66]['gungu'])
print('resNm', data[66]['resNm'])
print('rnum', data[66]['rnum'])
print('sido', data[66]['sido'])
print('yyyymm', data[66]['yyyymm'])
````

````python
data[2]['ForNum']
````

````python
ForNum = []
NatNum = []
addrCd = []
gungu = []
resNm = []
rnum = []
sido = []
yyyymm = []
for item in data:
    ForNum.append(item['ForNum'])
    NatNum.append(item['NatNum'])
    addrCd.append(item['addrCd'])
    gungu.append(item['gungu'])
    resNm.append(item['resNm'])
    rnum.append(item['rnum'])
    sido.append(item['sido'])
    yyyymm.append(item['yyyymm'])
    
print("ForNum: ", ForNum)
print("NatNum: ", NatNum)
print("addrCd: ", addrCd)
print("gungu: ", gungu)
print("resNm: ", resNm)
print("rnum: ", rnum)
print("sido: ", sido)
print("yyyymm: ", yyyymm)

# 1개의 키에 67개의 데이터가 있는 리스트를 묶어서 dict로 만들기

result = dict(ForNum=ForNum, NatNum=NatNum, addrCd=addrCd, gungu=gungu,
             resNm=resNm, rnum=rnum, sido=sido, yyyymm=yyyymm)
````

# key와 value를 수작업 없이 사용해서 정리하기
* dict에서 key는 반복문에 넣으면 key가 나옴
* keys(), values(), items()

````python
result2 = {}
for key in data[0]:
    temp = []
    for item in data:
        temp.append(item[key])
    result2[key] = temp  
result2
````

# 딕셔너리 함수를 이용해 만들기
* setdefault

````python
result3 = {}
for item in data:
    for key, value in item.items():
        result3.setdefault(key, []).append(value)
df3 = pd.DataFrame(result3)
df3
````

# collections.defaultdict 를 사용해서 더 빠르게

````python
from collections import defaultdict
````

````python
dic_result = defaultdict(list)
for item in data[:2]:
    for key, value in item.items():
        dic_result[key].append(value)
dic_result
````

````python
result4 = dict(dic_result)
result4
````
