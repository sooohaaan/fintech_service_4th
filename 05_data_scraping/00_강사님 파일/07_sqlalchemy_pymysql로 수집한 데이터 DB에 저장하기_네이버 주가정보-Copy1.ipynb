```python
dbid = os.getenv("dbid")
dbpw = os.getenv("dbpw")
host = os.getenv("host")
port = os.getenv("port")
```

```python
data = pd.read_csv("./data/상장기업정보.csv")
data
```

```python
engine = create_engine(f"mysql+pymysql://{dbid}:{dbpw}@{host}:{port}")
conn = engine.connect()
```

```python
result = conn.execute(text("show databases;"))
for row in result:
    print(row)
```

# 네이버 증권에서 주가 정보 수집하기

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 종목코드 가져오기
codes = data2['종목코드']
codes[:2]
```

```python
url = "https://finance.naver.com/item/main.naver?code=003480"
r = requests.get(url)
soup = bs(r.content, "lxml")
soup
```

```python
result = {}
for idx, dd in enumerate(soup.select("#middle dl.blind dd")[:-3]):
    
    if idx == 0:
        date = dd.get_text(strip=True).replace("\n", "").replace("  ", "").replace("년 ", "-").replace("월", "-").replace("일", "")[:10]
        result['날짜'] = date
    elif idx == 2:
        temp = dd.get_text().split()
        result[temp[0]] = temp[1]
        result["거래소"] = temp[2]
    elif idx == 3:
        temp = dd.get_text().split()
        result[temp[0]] = int(temp[1].replace(",", ""))
        if temp[3] == "상승":
            result[temp[2]] = int(temp[4])
        else:
            result[temp[2]] = -int(temp[4])
        if temp[5] == "플러스":
            result['증가율'] = temp[6]+"%"
        else:
            result['증가율'] = "-" + temp[6]+"%"
    else:
        if idx == 1 or idx == 11:
            temp = dd.get_text().split()
            result[temp[0]] = temp[1]
        else:
            print(dd)
            temp = dd.get_text().split()
            result[temp[0]] = int(temp[1].replace(",", ""))

print(result)
            

#     print(dd.get_text(strip=True))
```
