* while 반복문에 조건 추가

```python
result = {}

page = 1
while True:
    
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(divId='indsSclsCd', key='I21006', numOfRows=1000, pageNo=page, type='json', serviceKey=os.getenv("service_key")) 
    # divid 를 세부 업종으로, key를 치킨으로 수정
    r = requests.get(url, params=payload)
    print(r.url)
    print(r.status_code)
    response = r.json()
    
    
    # 전체 페이지수 구하기
    if response['body']['totalCount'] % 1000 == 0:
        total_pages = response['body']['totalCount']//1000
    else:
        total_pages = response['body']['totalCount']//1000 + 1

    # 현재 수집 페이지 출력
    print(f"{page}/{total_pages} 수집중    ", end="\r")
        
    # 종료 조건 주기
    if page > total_pages:
        break
    
    # 가게 정보 추출
    for item in response['body']['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    
    page += 1
    time.sleep(0.5)
    
df = pd.DataFrame(result) # 가장 마지막에 만들어야 하니까 반복문 밖으로 뺀다.
df
```

```python
result = {}

page = 1
total_pages = 1
while True:
    
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(divId='indsSclsCd', key='I21006', numOfRows=1000, pageNo=page, type='json', serviceKey=os.getenv("service_key")) 
    # divid 를 세부 업종으로, key를 치킨으로 수정
    r = requests.get(url, params=payload)
    print(r.url)
    print(r.status_code)
    response = r.json()
    
    
    # 전체 페이지수 구하기
    if total_pages == 1:
        if response['body']['totalCount'] % 1000 == 0:
            total_pages = response['body']['totalCount']//1000
        else:
            total_pages = response['body']['totalCount']//1000 + 1

    # 현재 수집 페이지 출력
    print(f"{page}/{total_pages} 수집중    ", end="\r")
        
    # 종료 조건 주기
    if page > total_pages:
        break
    
    # 가게 정보 추출
    for item in response['body']['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    
    page += 1
    time.sleep(0.5)
    
df = pd.DataFrame(result) # 가장 마지막에 만들어야 하니까 반복문 밖으로 뺀다.
df
```

```python
upjong_list = pd.read_excel("../API_매뉴얼_상가상권/소상공인시장진흥공단_상가(상권)정보_업종분류(2302)_및_연계표_v1.xlsx", header=1)
upjong_codes = upjong_list['소분류코드']
upjong_name = upjong_list['소분류명']
```

* 시작

```python
upjong_codes = upjong_list['소분류코드']
upjong_name = upjong_list['소분류명']

for code, name in zip(upjong_codes, upjong_name):
    print(f"{code}{name})")
        
    result = {}

    page = 1
    total_pages = 1
    while True:

        url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
        payload = dict(divId='indsSclsCd', key=code, numOfRows=1000, pageNo=page, type='json', serviceKey=os.getenv("service_key")) 
        # divid 를 세부 업종으로, key를 치킨으로 수정
        r = requests.get(url, params=payload)
        print(r.url)
        print(r.status_code)
        response = r.json()


        # 전체 페이지수 구하기
        if total_pages == 1:
            if response['body']['totalCount'] % 1000 == 0:
                total_pages = response['body']['totalCount']//1000
            else:
                total_pages = response['body']['totalCount']//1000 + 1

        # 종료 조건 주기
        if page > total_pages:
            break

        # 현재 수집 페이지 출력
        print(f"{page}/{total_pages} 수집중    ", end="\r")        

        # 가게 정보 추출
        for item in response['body']['items']:
            for key, value in item.items():
                result.setdefault(key, []).append(value)

        page += 1
        time.sleep(0.5)

    # dataframe
    df = pd.DataFrame(result)


    # csv 파일 생성
#     dir_name = "data"
#     if not os.path.exists(dir_name):
#         os.mkdir(dir_name)
#         print(f"{dir_name}을 생성했습니다.")
#     else:
#         print(f"{dir_name}이 이미 있습니다.")

    df.to_csv(f"./data/{name}_가게리스트.csv")


print("완료")
```

* 강사님 답안

```python
for idx, (code, name) in enumerate (zip(upjong_list['소분류코드'][:2], upjong_list['소분류명'])):
    print(code,name)
        
    result = {}

    page = 1
    total_pages = 1
    while True:

        url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
        payload = dict(divId='indsSclsCd', key=code, numOfRows=1000, pageNo=page, type='json', serviceKey=os.getenv("service_key")) 
        # divid 를 세부 업종으로, key를 치킨으로 수정
        r = requests.get(url, params=payload)
#         print(r.url)
#         print(r.status_code)
        response = r.json()


        # 전체 페이지수 구하기
        if total_pages == 1:
            if response['body']['totalCount'] % 1000 == 0:
                total_pages = response['body']['totalCount']//1000
            else:
                total_pages = response['body']['totalCount']//1000 + 1

        # 종료 조건 주기
        if page > total_pages:
            break

        # 현재 수집 페이지 출력
        print(f"{idx}/{len(upjong_list['소분류코드'])}_{name}_{page}/{total_pages}수집중", end="\r")        

        # 가게 정보 추출
        for item in response['body']['items']:
            for key, value in item.items():
                result.setdefault(key, []).append(value)

        page += 1
        time.sleep(0.5)

    # dataframe
    df = pd.DataFrame(result)


    # csv 파일 생성
    dir_name = "shops_data"
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
        print(f"{dir_name}을 생성했습니다.")
    else:
        print(f"{dir_name}이 이미 있습니다.")

    df.to_csv(f"../shops_data/{name}_가게리스트.csv",encoding="utf-8-sig")


print("완료")
```
