## 1.5. DB 접속

### (1) DB 접속

```python
# mysql에 접속할 때, pymysql을 이용할 것이다.
# create_engine은 워크벤치의 접속 카드 만드는 것과 같다.
# hostname:port
# conn => mysql의 쿼리창
engine = create_engine("mysql+pymysql://root:1234@127.0.0.1:3306")
conn = engine.connect()
```

### (2) DB 접속 확인

```python
conn.execute(text("show databases;"))
```

### (3) DB 목록 조회

```python
# DB 목록 조회
result = conn.execute(text("show databases;"))
for row in result:
    print(row)
```
