# XML로 가져오기
* xml: html처럼 태그 기반으로 자료를 저장한 포맷
* xml parser를 통해서 str을 xml로 변환하는 작업이 필요
* xml로 변환이 되면 태그 기반으로 자료를 찾아서 정리
* 태그에서 자료를 추출할 때는 beautifulsoup이라는 라이브러리를 이용

* find, find_all 함수: xml, html에서 태그 기반으로 내용을 찾음  ===> 가장 많이 사용
* (여려개를 찾을 때: find_all, select)
* select, select_one 함수: xml, html에서 css selector기반으로 내용을 찾음  ===> 가장 많이 사용
* find_all, select는 해당 태그나 css selector를 가진 부분을 모두 찾아서 list로 반환
* find, select_one은 여러 태그나 css selector 중에서 가장 먼저 나오는 태그/selector를 한 개만 찾아줌
* 찾아온 태그 .name: 태그 이름 반환
* 찾아온 태그 .get_text(): 태그 안쪽의 텍스트 반환

````python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
</body>
</html>
"""
````

````python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
````

````python
# b 태그를 찾고 싶다.
soup.find_all("b") # a 태그를 다 찾아준다.
````

* find, find_all 태그 기반으로 찾기
    * find는 html이나 xml에서 가장 위에 있는 태그 1개만 찾아줌
    * find_all은 html 내부에 있는 모든 태그를 찾아서 list로 만들어줌

````python
soup.find("a")
````

````python
# a 태그를 찾고 싶다.
soup.find_all("a") # a 태그를 다 찾아준다.
````

````python
# find_all 은 태그를 찾아서 list로 반환, for 반복문 필요
# 인덱싱이 가능하다.
soup.find_all("a")[::-1]
````

````python
type(soup.find_all("a")) # type이 list가 아니다. beautifulsoup 객체이다.
````

* find로 찾은 태그의 이름 출력 .name => dict의 key출력
* find로 찾은 태그 안의 내용 출력 .get_text() => dict의 value출력

````python
# html에 있는 모든 태그의 이름 출력
for tag in soup.find_all():
    print(tag.name) # key를 뽑는 과정이라고 보면 된다.
````

````python
# html에 있는 모든 태그의 이름 출력
for tag in soup.find_all():
    print(tag.name, tag.get_text()) # value를 뽑는 과정이라고 보면 된다.
````

# xml에서 데이터추출하기

````python
url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
payload = dict(divId='indsLclsCd', key='I2', numOfRows=2, pageNo=1, type='xml', serviceKey=os.getenv("service_key")) 
r = requests.get(url, params=payload)
print(r.url)
print(r.status_code)
soup = bs(r.content, "xml")

result = {}
for item in soup.find_all('item'): # 리스트가 없어졌다.
    for tags in item.find_all(recursive=False):
        result.setdefault(tags.name, []).append(tags.get_text())
df = pd.DataFrame(result)
df
````

* 내가 한 코드

````python
page = 1
total_pages = 1
result = {}
while True:

    # xml에서 데이터추출코드 (API 데이터 요청 및 soup 객체 생성)
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(divId='indsLclsCd', key='I2', numOfRows=1000, pageNo=page[:10] type='xml', serviceKey=os.getenv("service_key")) 
    r = requests.get(url, params=payload)
#     print(r.url)
#     print(r.status_code)
    soup = bs(r.content, "xml")

    # page수 계산
    if total_pages == 1:
        if int(soup.find("totalCount").get_text()) % 1000 == 0:
            total_pages = int(soup.find("totalCount").get_text()) // 1000
        else:
            total_pages = int(soup.find("totalCount").get_text()) // 1000 + 1

    # 데이터 수집 중단
    if page > total_pages:
        break

    # 현재 페이지 위치
    print(f"{page}/{total_pages}수집중", end="\r")
    
    for item in soup.find_all('item'): # 리스트가 없어졌다.
        for tags in item.find_all(recursive=False):
            result.setdefault(tags.name, []).append(tags.get_text())

    page += 1


df = pd.DataFrame(result)
df
````

# 모든 카테고리(업종)데이터 수집하고 업종별로 저장하기

````python
upjong = pd.read_excel("./API_매뉴얼_상가상권/소상공인시장진흥공단_상가(상권)정보_업종분류(2302)_및_연계표_v1.xlsx", header=1)
````

````python
codes = upjong['소분류코드']
names = upjong['소분류명']
````

````python
for code, name in zip(codes, names):
    print(code,name)
````

````python
for idx, (code, name) in enumerate (zip(codes[:3], names)):
#     print(code,name)
    result = {}
    page = 1
    total_pages = 1

    while True:

        # 종료 조건
        if page > total_pages:
            break

        url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
        payload = dict(divId='indsLclsCd', key='I2', numOfRows=1000, pageNo=1, type='xml', serviceKey=os.getenv("service_key")) 
        r = requests.get(url, params=payload)
#         print(r.url)
#         print(r.status_code)
        soup = bs(r.content, "xml")

        # 한국어 컬럼 추출하기
        kor_cols = soup.find("columns").get_text().split(",")

        # 전체 페이지 구하기
        if total_pages == 1:
            if int(soup.find("totalCount").get_text()) % 1000 == 0:
                total_pages = int(soup.find("totalCount").get_text()) // 1000
            else:
                total_pages = int(soup.find("totalCount").get_text()) // 1000 + 1

        print(f"{idx}/{len(names)-1}카테고리 데이터 {page}/{total_pages}", end="\r")        
                
        
        # 데이터 추출해서 result에 모으기
        for item in soup.find_all('item'): # 리스트가 없어졌다.
            for tags in item.find_all(recursive=False):
                result.setdefault(tags.name, []).append(tags.get_text())
            
        page += 1
        time.sleep(0.3)
    
    # 폴더 자동으로 생성하기
    dir_name = "shops_data_xml"
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)

    df = pd.DataFrame(result)
    df.to_csv(f"./{dir_name}/{name}_데이터.csv", encoding="utf-8-sig")
````

````python
page = 1
total_pages = 1
result = {}
while True:

    # xml에서 데이터추출코드 (API 데이터 요청 및 soup 객체 생성)
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(divId='indsLclsCd', key='I2', numOfRows=1000, pageNo=page, type='xml', serviceKey=os.getenv("service_key")) 
    r = requests.get(url, params=payload)
#     print(r.url)
#     print(r.status_code)
    soup = bs(r.content, "xml")

    # page수 계산
    if total_pages == 1:
        if int(soup.find("totalCount").get_text()) % 1000 == 0:
            total_pages = int(soup.find("totalCount").get_text()) // 1000
        else:
            total_pages = int(soup.find("totalCount").get_text()) // 1000 + 1

    # 데이터 수집 중단
    if page > total_pages:
        break

    # 현재 페이지 위치
    print(f"{page}/{total_pages}수집중", end="\r")
    
    for item in soup.find_all('item'): # 리스트가 없어졌다.
        for tags in item.find_all(recursive=False):
            result.setdefault(tags.name, []).append(tags.get_text())

    page += 1
    
    # csv 파일 생성
    df.to_csv(f"./{}_가게리스트.csv", encoding="utf-8-sig")
    
print("완료")
````
