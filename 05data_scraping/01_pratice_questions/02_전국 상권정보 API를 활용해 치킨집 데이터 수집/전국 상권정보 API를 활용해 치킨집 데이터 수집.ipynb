# 3. XML

### XML로 가져오기
* xml: html처럼 태그 기반으로 자료를 저장한 포맷
* xml parser를 통해서 str을 xml로 변환하는 작업이 필요
* xml로 변환이 되면 태그 기반으로 자료를 찾아서 정리
* 태그에서 자료를 추출할 때는 beautifulsoup이라는 라이브러리를 이용
