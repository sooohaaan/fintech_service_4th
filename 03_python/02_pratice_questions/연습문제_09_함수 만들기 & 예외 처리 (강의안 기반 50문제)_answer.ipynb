## 31번 문제

```python
# 함수(1차 시도)
def multi_divide(*args):
    try:
        result = args[0]
        for num in args[1:]:
            result = result / num
    except Exception as e:
        print(e)
    else:
        print(result)
        print("정상처리")
```

```python
# 출력
multi_divide(10,2)
```

```python
# 함수(2차 시도)
def multi_divide(*args):
    try:
        result = args[0]
        for num in args[1:]:
            result = result / num
            print(result)
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
    else:
        print("정상처리")
```

```python
# 출력
multi_divide(10,2)
```

```python
# 정답
a=10
b=2
try:
    print(a/b)
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다')
else:
    print('정상처리')
```

## 37번 문제

```python
# 함수(1차 시도)
def range_value(nums):
    for i in nums:
        a = min(i)
        b = max(i)
    return b - a
```

```python
# 함수(2차 시도)
def range_value(nums):
    a = min(nums)
    b = max(nums)
    return b - a
```

```python
# 함수(정답)
def range_value(nums):
    return max(nums) - min(nums)
```

```python
print(range_value([3,10, 1, 7]))
```
