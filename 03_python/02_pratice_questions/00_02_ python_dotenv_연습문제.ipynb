## 7번

```python
# env 파일을 읽어서 환경 변수로 등록한다.
from dotenv import load_dotenv
```

```python
# 패스라이브러리에서 패스를 가져온다.
from pathlib import Path
```

```python
# 현재 경로 찾기(1)
print(__file__)
```

```python
# 현재 경로 찾기(2)
print(os.getcwd())
```

```python
# 파일의 위치를 기준으로 env 파일 찾아 불러온다.
BASE_DIR = Path(__file__).resolve().parent
```

```python
# 사용자의 위치를 기준으로 env 파일 찾아 불러온다.
BASE_DIR = Path(os.getcwd()).resolve().parent
```

```python
# 알아낸 경로에 .env를 붙여준다.
load_dotenv(BASE_DIR / '.env')
```
