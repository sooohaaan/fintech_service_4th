## 문제 2번

### 자기소개

* 1안

```python
# **<1안>**

# 딕셔너리 자료형 s
# dict함수 사용
s = dict(사는_곳='서울', 성별='여자', 이름='시은', 나이=11, 혈액형='b')

# dict 함수를 사용하여 딕셔너리 자료형을 만들 때, '사는 곳'키의 띄어쓰기 적용이 불가하다. 
# 때문에 띄어쓰기가 필요한 영역에 '_'를 사용 후 반복문과 replace 함수를 사용하여 '_'를 ' ' 으로 전환한다. 
result = {}
for k, v in s.items():
    new_k = k.replace('_', ' ')
    result[new_k] = v

# 출력 코드
print(f"""내가 살고 있는 곳은 {result['사는 곳']}입니다.
나의 성별은 {result['성별']} 이고 이름은 {result['이름']} 입니다.
나이는 {result['나이']}살 이고 혈액형은 {result['혈액형']} 형입니다.""")
```

* 2안

```python
# **<2안>**

# 딕셔너리 자료형 s
s = {'사는 곳':'서울', '성별':'여자', '이름':'시은', '나이':11, '혈액형':'b'}

# 출력 코드
print(f"""내가 살고 있는 곳은 {s['사는 곳']}입니다.
나의 성별은 {s['성별']} 이고 이름은 {s['이름']} 입니다.
나이는 {s['나이']}살 이고 혈액형은 {s['혈액형']} 형입니다.""")
```

* 정답

```python
# 정답 1

s = dict([('사는 곳','서울'),('성별','여자'),('이름','시은'),('나이',11), ('혈액형','b')])
s
```

```python
# 정답 2

print(f"내가 살고 있는 곳은 {s['사는 곳']}입니다.")
print(f"나의 성별은 {s['성별']}이고 이름은 {s['이름']}입니다.")
print(f"나이는 {s['나이']} 살 이고 혈액형은 {s['혈액형']} 형입니다.")
```

## 문제 3번

### 딕셔너리에 추가하기

* 1안

```python
# 딕셔너리 자료형
s = {'사는 곳':'인천 송도','성별':'남자','이름':'우진'}

# 입력 코드
s['나이']=input('나이는 몇 살인가요?')
s['혈액형']=input('혈액형은 무엇인가요?')

# 출력 코드
print(f"""나의 성별은 {s['성별']} 이고 이름은 {s['이름']} 입니다.
나이는 {s['나이']} 살 이고 혈액형은 {s['혈액형']} 형입니다.""")
```

* 2안

```python
# 딕셔너리 자료형
# dict 함수 사용
s=dict(사는_곳='인천 송도', 성별='남자', 이름='우진')


# dict 함수를 사용하여 딕셔너리 자료형을 만들 때, '사는 곳'키의 띄어쓰기 적용이 불가하다. 
# 때문에 띄어쓰기가 필요한 영역에 '_'를 사용 후 반복문과 replace 함수를 사용하여 '_'를 ' ' 으로 전환한다. 
result = {}
for k, v in s.items():
    new_k = k.replace('_', ' ')
    result[new_k] = v

# 입력 코드
result['나이']=int(input('나이는 몇 살인가요?'))
result['혈액형']=input('혈액형은 무엇인가요?').upper() # 입력 문자를 대문자로 전환

# 출력 코드
print(f"""나의 성별은 {result['성별']} 이고 이름은 {result['이름']} 입니다.
나이는 {result['나이']} 살 이고 혈액형은 {result['혈액형']} 형입니다.""")
```

* 3안

```python
# 딕셔너리 자료형
# dict 함수 사용
s = {'사는 곳':'인천 송도', '성별':'남자', '이름':'우진'}


# 입력 코드
s['나이']=int(input('나이는 몇 살인가요?'))
s['혈액형']=input('혈액형은 무엇인가요?').upper()


# 출력 코드
print(f"""나의 성별은 {s['성별']} 이고 이름은 {s['이름']} 입니다.
나이는 {s['나이']} 살 이고 혈액형은 {s['혈액형']} 형입니다.""")
```

## 문제 6번

### 딕셔너리와 if 조건문의 만남

* 1안

```python
# 코드
store = {'새우칩':1500, '옥수수칩':1800, '콜라':700, '양파칩':1300}

# 입력 코드

store=input('물품을 입력하면 가격을 알려드립니다.\n물품 이름은?')

# 출력 코드
print(f"""{store}의 가격은 {store} 입니다.
{store}는 등록되어 있지 않습니다.""")
```

```python
# 코드
store = {'새우칩':1500, '옥수수칩':1800, '콜라':700, '양파칩':1300}

# 입력 코드
print('물품을 입력하면 가격을 알려드립니다.')
input_name=input('물품 이름은?')

# 출력 코드
if input_name in store:
    print(f"""{input_name}의 가격은 {store[input_name]} 입니다.""")
else:
    print(f"""{input_name}는 등록되어 있지 않습니다.""")
```

## 문제 7번

### 딕셔너리의 값으로 리스트 사용하기

* 1안

```python
# 반복문 미사용

# store 딕셔너리
store = {'새우칩':[1500,3], '옥수수칩':[1800,2],'콜라':[700,4],'양파칩':[1300,1]}

# key, value 구분
name_list = list(store.keys()) # ['새우칩', '옥수수칩', '콜라', '양파칩']
value_list = list(store.values()) # [[1500, 3], [1800, 2], [700, 4], [1300, 1]]

# 출력 코드
print(f"1 번 물품은 {name_list[0]} 이고 가격은 {value_list[0][0]} 원이고 수량은 {value_list[0][1]} 개입니다.")
print(f"2 번 물품은 {name_list[1]} 이고 가격은 {value_list[1][0]} 원이고 수량은 {value_list[1][1]} 개입니다.")
print(f"3 번 물품은 {name_list[2]} 이고 가격은 {value_list[2][0]} 원이고 수량은 {value_list[2][1]} 개입니다.")
print(f"4 번 물품은 {name_list[3]} 이고 가격은 {value_list[3][0]} 원이고 수량은 {value_list[3][1]} 개입니다.")


# 물건 종류
total_type_count = len(store)

# 총 물건 개수
total_qty = store['새우칩'][1] + store['옥수수칩'][1] + store['콜라'][1] + store['양파칩'][1]

# 총 가격
total_price = store['새우칩'][0] * store['새우칩'][1] + store['옥수수칩'][0] * store['옥수수칩'][1] + store['콜라'][0] * store['콜라'][1] + store['양파칩'][0] * store['양파칩'][1]

# 평균 가격
total_avg = total_price / total_qty

# 출력 코드
print(f"우리 가게는 {total_type_count} 개의 물건 종류가 있습니다.")
print(f"우리 가게 모든 물건 수량은 {total_qty} 개 있습니다.")
print(f"우리 가게 모든 물건들의 가격 총합은 {total_price} 원입니다.")
print(f"우리 가게 모든 물건들의 평균 가격은 {total_avg} 원입니다.")
```

## 문제 8번

### 딕셔너리의 값으로 딕셔너리 사용하기

* 1안

```python
# 
student = {'우진':{'코딩':95, '수학':99}, '시은':{'과학':100,'코딩':99},'메이킷':{'영어':85,'코딩':100}}

# 출력 코드
print(f"우진이의 코딩 점수는? {student['우진']['코딩']}")
print(f"시은이의 과학 점수는? {student['시은']['과학']}")
print(f"메이킷의 영어 점수는? {student['메이킷']['영어']}")
```

```python
# 딕셔너리 데이터(중첩 딕셔너리)
student_result = dict(우진=dict(코딩=95, 수학=99), 시은=dict(과학=100,코딩=99), 메이킷=dict(영어=85, 코딩=100))

# 반복문 사용
for name in student_result:

    subject_list = list((student_result[name]).keys())
    
    first_subject = subject_list[0]
    
    scores = student_result[name][first_subject]
    
    print(f"{name}이의 {first_subject} 점수는? {scores}")
```
