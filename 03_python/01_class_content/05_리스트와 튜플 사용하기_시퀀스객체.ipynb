# 튜플 tuple
* 튜플도 리스트처럼 모든 자료형을 만들 수 있다.
* 순서가 있는 자료형, 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 안됨
* ()안에 자료를 넣어서 만든다.
* 자료 1개만 넣어서 만들 때는 반드시 (자료, ) 형식이어야 한다.
* 튜플 생성 시 자료가 2개 이상인 경우 () 생갹 가능
* 빈 튜플 생성가능

```python
t = (1,2,3)
t
```

```python
t2 = ()
print(t2, type(t2))
```

```python
t[0]
```

```python
t[:2]
```

```python
t[::-1]
```

```python
# sort는 요소를 변경하기 때문에 작동 안됨.
t.sort()
```

```python
t = sorted(t, reverse=True)
```

```python
t
```

```python
t = tuple(t)
t
```

```python
sorted(t) # 리스트로 바꾼다.
```

# 시퀀스 객체의 특징
* 문자열, list, tuple
* 순서가 있는 자료형
* 인덱싱, 슬라이싱이 가능한 자료
* 반복문에 넣어서 자료를 하나씩 꺼내는 것이 가능
* 시퀀스 객체 + 시퀀스 객체 = 합쳐짐
* 시퀀스 객체 * 숫자 = 숫자만큼 반복 복사
* 객체 = 포장

```python
s = "문자열"
l1 = [1,2,3]
l2 = [4,5,6]
t1 = (1,2,3)
t2 = (4,5,6)
```

```python
s + s
```

```python
l1 + l2
```

```python
(l1 + l2) * 3
```

```python
t1 + t2
```

```python
(t1 + t2) * 3
```

```python
((t1 + t2) * 3)[::-1]
```

```python
for item in s:
    print(item, end=" ")
```

```python
for item in l1:
    print(item, end=" ")
```

```python
for item in ((t1 + t2) * 3)[::-1]:
    print(item)
```
