```python
import json
import pandas as pd
```

```python
with open("./data/com_kbstar_kbbank.json", encoding="utf-8") as file:
    raw_data2 = file.read()
    bank_data = json.loads(raw_data2)
```

```python
 len(bank_data)
```

```python
 = bank_data[0]
```

```python
result = {}
for test1 in bank_data:
    for key, value in test1.items():
        if key == "별점":
#             print(value, type(value))
            value = int(value)
        elif key == "리뷰":
            value = value.replace("?", " ").replace(".", "").replace("ㅋ", "")
        elif key == "개발자 응답":
            if value != None:
                value = value.replace("?", " ").replace(".", "").replace("ㅋ", "") 
            
        result.setdefault(key, []).append(value)
df = pd.DataFrame(result)
df
```

# 전국 치킨집 리스트 데이터 정리하기

```python
with open("./data/전국치킨집리스트.json") as file:
    raw_data = file.read()
    data = json.loads(raw_data)
```

```python
result2 = {}
for page in data:
    for shop in page['body']['items']:
        for key, value in shop.items():
            result2.setdefault(key, []).append(value)
df2 = pd.DataFrame(result2)
df2
```
