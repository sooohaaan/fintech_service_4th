# if else 조건문

## 1. 조건문? 특정 조건을 판별해서 작업을 분기시키는 문법
```python
if 조건식:
    조건이 참일 때 실행할 코드
else:
    조건이 거짓일 때 실행할 코드
```

````python
x = 12
````

````python
# x가 10인지 아닌지 판변하는 조건문
if x == 10 or x % 4 == 0:
    print(f"{x}는, 10입니다.")
    print(f"{x}는, 10이 아닌 {x}입니다.")
else:
    print(f"{x}는, 10이 아닙니다.")
````

* 숫자를 입력받아 홀수인지 짝수인지 판단하는 프로그램

````python
x = input("숫자를 입력하세요: ")
print(x, type(x))
if int(x) % 2 == 0:
    print(f"입력한 숫자{x}는 짝수입니다.")
else:
    print(f"입력한 숫자{x}는 홀수입니다.")
    
````

* if문의 조건을 여러개 주기

````python
# 숫자를 입력받아서 숫자가 1-10사이의 숫자인지 판별하고
# 홀수인지 짝수인지 판별해서
# 입력받은 숫자 {x}는 1-10 사이의 숫자이고/숫자가 아니고 홀수/짝수입니다. 
x = input("숫자를 입력해주세요: ")
x = int(x)
# x가 1-10 사이의 숫자이고 짝수인 경우
if 1 <= x <= 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자이고 짝수입니다.")
else:
    print(f"{x}는 1-10 사이의 숫자가 아니고 홀수입니다.")
    
````

## 2. 다중 조건문 사용하기 elif
``` python
if 조건:
    조건이 참일 때 실행할 코드
elif 조건2:
    조건2가 참일 때 실행할 코드
elif 조건3:
    조건3이 참일 때 실행할 코드
else:
    위의 조건1, 2, 3이 모두 거짓일 때 실행할 코드
```
    

````python
# 숫자를 입력받아서 숫자가 1-10사이의 숫자인지 판별하고
# 홀수인지 짝수인지 판별해서
# 입력받은 숫자 {x}는 1-10 사이의 숫자이고/숫자가 아니고 홀수/짝수입니다. 
x = input("숫자를 입력해주세요: ")
x = int(x)
# x가 1-10 사이의 숫자이고 짝수인 경우
if 1 <= x <= 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자이고 짝수입니다.")
elif 1 <= x <= 10 and x % 2 == 1:
    print(f"{x}는 1-10 사이의 숫자이고 홀수입니다.")
elif x > 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자가 아니고 짝수입니다.")    
else:
    print(f"{x}는 1-10 사이의 숫자가 아니고 홀수입니다.")
    
````

### else 없이 if와 elif로만 처리하는 경우

````python
x = input("숫자를 입력해주세요: ")
x = int(x)
# x가 1-10 사이의 숫자이고 짝수인 경우
if 1 <= x <= 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자이고 짝수입니다.")
elif 1 <= x <= 10 and x % 2 == 1:
    print(f"{x}는 1-10 사이의 숫자이고 홀수입니다.")
elif x > 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자가 아니고 짝수입니다.")    
elif x > 10 and x % 2 == 1:
    print(f"{x}는 1-10 사이의 숫자가 아니고 홀수입니다.")
````

### else없이 if문이 참일 때만 일하도록 하는 경우

````python
x = input("숫자를 입력해주세요: ")
x = int(x)
# x가 1-10 사이의 숫자이고 짝수인 경우
if 1 <= x <= 10 and x % 2 == 0:
    print(f"{x}는 1-10 사이의 숫자이고 짝수입니다.")
````

* 점수를 입력받아서 81-100은 A등급, 61-80은 B등급 41-60은 C등급 21-40은 D등급 0-20은 F등급 입니다. 를 출력하는 조건문을 만들어보세요.

````python
score = int(input("점수를 입력하세요: "))
print(score, type(score))
if 81 <= score <= 100:
    print(f"{score}는 A등급입니다.")
elif 61 <= score <= 80:
    print(f"{score}는 B등급입니다.")
elif 41 <= score <= 60:
    print(f"{score}는 C등급입니다.")
elif 21 <= score <= 40:
    print(f"{score}는 D등급입니다.")
else:
    print(f"{score}는 F등급입니다.")
````

* 입력값이 0-100사이인지 판별하고 범위를 벗어나면 0-100사이의 숫자를 다시 입력해주세요.

````python
score = int(input("점수를 입력하세요: "))
print(score, type(score))
if 0 <= score <= 100:
    if 81 <= score <= 100:
        print(f"{score}는 A등급입니다.")
    elif 61 <= score <= 80:
        print(f"{score}는 B등급입니다.")
    elif 41 <= score <= 60:
        print(f"{score}는 C등급입니다.")
    elif 21 <= score <= 40:
        print(f"{score}는 D등급입니다.")
    else:
        print(f"{score}는 F등급입니다.")
else:
    print(f"""현재 입력된 값{score}입니다.
0에서 100사이의 숫자를 다시 입력해 주세요.""")
````

* 1-50 사이의 숫자를 입력받아서 3의 배수면 3의 배수입니다.
* 3의 배수가 아니면 3의 배수가 아닙니다.
  * 입력받기, 정수로 변환하기
  * 1-50 사이의 숫자인가?
  * 3의 배수인가? % 3 == 0
  * 결과 출력

````python
num = int(input("정수로 숫자를 입력하세요: "))
# print(num, type(num))
if 1 <= num <= 50:
#     print(f"{num}은 1-50사이의 숫자입니다")
    if num % 3 == 0:
        print(f"{num}은 3의 배수입니다.")
    else:
        print(f"{num}은 3의 배수가 아닙니다.")
else:
    print("1-50 사이의 정수로 입력해 주세요.")
````

# 3.조건 표현식
```
참일 때 실행할 코드 if 조건 else 거짓일 때 실행할 코드
```

````python
# 일반적인 if else문으로 홀수, 짝수 구분하는 코드
num = int(input("숫자를 입력하세요"))
# print(num, type(num))
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
````

````python
# 조건표현식으로 홀수, 짝수 구분하는 코드
num = int(input("숫자를 입력하세요"))
# print(num, type(num))
print("짝수") if num % 2 == 0 else print("홀수")    
````
