# 서식지정자, format 메서드, f-string
* 문자열로만 입력해야 하는 상황에서 같은 위치세 문자를 여러 번 바꿔야 할 때 사용
* 문자의 자릿수 맞춤, 위치 정렬에 사용

## 1. 서식지정자 %d, %s, %f

### 1) 문자를 입력할 때 %s (sting)

```python
print("이 수업의 이름은 %s 입니다." % "국어")
```

```python
for name in ['국어', '수학', '사회', '영어']:
    print("이 수업의 이름은 %s 입니다." % name)
```

### 2) 정수를 입력할 때 %d

```python
print("올해는 %d년 입니다." % 2026)
```

```python
print("올해는 %d년 입니다." % 2026.55)
```

### 3) 실수를 입력할 때 %f 기본 소숫점 6자리로 표시

```python
print("사람의 정상 체온은 약 %f도 입니다." % 36.5)
```

* 실수를 입력할 때 소숫점 자리 수를 정하고 싶을 때 %자릿수f

```python
print("사람의 정상 체온은 약 %.1f도 입니다." % 36.5)
```

```python
print("사람의 정상 체온은 약 %.2f도 입니다." % 36.5)
```

## 4) 서식지정자로 문자열 길이, 정렬 %길이s

```python
print("%3s" % "01")
```

```python
"%3s" % "01"
```

```python
"%5s" % "01"
```

```python
print(150000)
print(1500)
```

```python
print("가격은 %10d 입니다." % 150000) 
print("가격은 %10d 입니다." % 1500)
```

### 5) 여러 항목 동시에 놓기

```python
print("오늘은 %4d년 %2d월 %2d일 입니다." % (2026, 1, 8))
```

```python
print("오늘은 %4d년 %02d월 %02d일 입니다." % (2026, 1, 8))
```

```python
company = ['삼성전자', 'LG전자', "현대차", "대한항공", "하나투어"]
stock = [122000, 150000, 32000, 24000, 86600]
```

```python
for com, st in zip(company, stock):
    print("%s의 주가는 %d 입니다." % (com, st))
```

## 2. format 메서드 사용하기
* "문자열 {인덱스번호} 문자열 끝".format(값)

```python
print("{}의 주가는 {}입니다.".format('삼성전자', 140000))
```

```python
print("{}의 주가는 {}입니다.".format('삼성전자'))
```

```python
print("{0}의 주가는 {0}입니다.".format('삼성전자'))
```

```python
print("{0}의 주가는 {1}입니다.".format('삼성전자', 140000))
```

```python
print("""{0}의 주가는 {1}원 입니다.
{0}의 상한가는 {2}입니다.
{0}의 하한가는 {3}입니다.""".format("삼성전자", 140000, 143000, 139000))
```

```python
for com, st in zip(company, stock):
    print("{}의 주가는 {} 입니다.".format(com, st))
```

## f-string으로 값 넣기

```python
print(company)
print(stock)
```

```python
print(f"{company[0]}의 주가는 {stock[0]}입니다.")
```

```python
for com, st in zip(company, stock):
    print(f"{com}의 주가는 {st} 입니다.")
```

# 3. f-sting으로 문자열 정렬, 자릿수 지정해 출력하기

```python
year = 2026
month = 1
day = 8
print(f"오늘은 {year}년 {month}월 {day}일 입니다.")
```

```python
# print("오늘은 %4d년 %02d월 %02d일 입니다." % (2026, 1, 8))
print(f"오늘은 {year}년 {month:02d}월 {day:02d}일 입니다.")
```

```python
url = "https://finance.naver.com/sise/sise_trans_style.naver?sosok="
```

```python
for i in range(1, 11):
    print(f"{url}{i:03d}")
```
