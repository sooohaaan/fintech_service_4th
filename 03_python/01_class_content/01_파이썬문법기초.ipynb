# print문의 옵션
* sep: 구분자 - print문 안에 여러개의 자료를 출력할 때 자료를 구분해주는 문자
* 기본값은 " " 공백 (공백도 문자다)

```python
print()
```

```python
print("이것은", '파이썬', '입니다.')
```

```python
print("이것은", '파이썬', '입니다.', sep="_")
print("이것은", '파이썬', '입니다.', sep=" ")
print("이것은", '파이썬', '입니다.', sep="")
```

```python
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
```

* end 옵션 - print가 끝나고 실행될 옵션
* 기본값 = \n - 이스케이프 문자 new line 새줄

```python
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
```

```python
# end 값을 공백 " " 으로 준 경우
print("헐크", '아이언맨', '스파이더맨', sep=" vs ", end=" ")
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
```

```python
# end 값을 공백 "\t" (tap)으로 준 경우 (이스케이프 문자 \t)
print("헐크", '아이언맨', '스파이더맨', sep=" vs ", end="\t")
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
```

```python
# end 값을 공백 "\t" (tap)으로 준 경우
print("헐크", '아이언맨', '스파이더맨', sep=" vs ", end=" >> ")
print("헐크", '아이언맨', '스파이더맨', sep=" vs ")
```

```python
for i in range(100):
    print(i, end=" ")
```

```python
import time
for i in range(1, 101):
#     print(i, "/" 100, end="\r")
#     print(i, "/", 100)
    time.sleep(0.5)
```

```python
# end 값을 공백 "\r" (carrige return)으로 준 경우
print("헐크", '아이언맨', '스파이더맨', sep=" vs ", end="\r")
print("스파이더맨", '아이언맨   ', sep=" vs ")
```

```python
# end 값을 공백 "\n" (new line)으로 준 경우
print("헐크", '아이언맨', '스파이더맨', sep=" vs ", end="\n\n\n")
print("스파이더맨", '아이언맨', '헐크헐크', sep=" vs ")
```

```python
# end 값을 공백 "\n" (new line)으로 준 경우
print("헐크", 1, 3.14, sep="-", end="\n\n\n")
print("스파이더맨", '아이언맨', '헐크헐크', sep=" vs ")
```

# 변수
* 자료를 담는 택배상자 ("="는 "같다"가 아니라 "넣는다"이다)
* 택배 상자는 1개만 넣을 수 있다.
* 변수는 영문과 숫자를 조합해서 만들 수 있다.
* 변수명의 시작은 반드시 영문자로 시작 or _로 시작
* 변수명이 숫자로 시작하거나 -등의 특수문자로 시작하면 오류
* 예약어(이미 파이썬에서 사용하고있는 함수명, 문법용 단어 ex) for, list, if, id)는 변수명으로 사용할 수 없다.
  * 주의!! 예약어를 써도 파이썬에서는 변수가 만들어진다.
  * 그러므로 예약어를 잘 기억해두자!!
  * 변수명을 입력했을 때 녹색으로 표시되면 예약어(문법용 단어)이다.
