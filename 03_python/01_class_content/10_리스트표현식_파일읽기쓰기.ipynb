# 반복문 + if문의 리스트 표현식
* 필터링
```python
for...if 조건(for문이 앞에, if 조건문이 뒤에, else 없음)
```
*if else + for문의 경우
``` python
참일 때 if 조건 else 거짓일 때 for...(조건 표현식이 앞, for문은 뒤에, if else 함께)
```

## 1) 조건을 이용해 필터링 하기(if만 쓸 때)
* 1-10 중 짝수만 모으기

````python
# 일반 for + if문
evens = []
for num in range(1,11):
    if num % 2 == 0:
        evens.append(num)
evens
````

````python
# 일반 for + if문
evens2 = [num for num in range(1,11) if num % 2 == 0]
evens2
````

## 2) 조건에 따라서 값을 바꾸기 (if ~ else)

* 1-10중 짝수면 짝수 홀수면 홀수를 출력

````python
# for + if
lables = []
for num in range(1,11):
    if num % 2 == 0:
        lables.append("짝수")
    else:
        lables.append("홀수")
lables
````

````python
# 리스트표현식 / 꼭 알고 있어야 한다.
# 조건표현식 + for문
lables2 = ["짝수" if num % 2 == 0  else "홀수" for num in range(1,11)]
lables2
````

````python
print(lables2, type(lables2))
````

# 제네레이터 표현식
* 리스트표현식과 문법 동일
* [] 아닌 ()로 만듬
* 1회용
* 만들고 나서 반복문으로 내용을 꺼내면 사라짐
* 오직 순차 접근만 가능, 인덱싱, 슬리이싱 불가
* 데이터가 너무 크거나, 한 번만 쓸 데이터인 경우 사용

````python
# 1-10까지 숫자중 짝수인 것만 모은 리스트를 제네레이터로
evens_g = (num for num in range(1, 11) if num % 2 == 0)
evens_g
````

* 끝까지 꺼내고 나면 사라짐

````python
# next() 제네레이터에서 자료를 하나씩 꺼낼 때 사용
next(evens_g)
````

# 리스트와 제네레이터 크기 비교

````python
!pip install pympler
````

````python
import sys
from pympler import asizeof
import psutil, os

# 메모리 사용량 확인 함수
def show_memory(msg):
    process = psutil.Process(os.getpid())
    print(f"{msg}: {process.memory_info().rss / 1024**2:.2f} MB")

# 비교 대상 생성
list_var = [x for x in range(1, 1_000_000)]   # 리스트 내포
gen_var  = (x for x in range(1, 1_000_000))   # 제너레이터 표현식

print("==== sys.getsizeof() 결과 ====")
print("리스트:", sys.getsizeof(list_var))
print("제너레이터:", sys.getsizeof(gen_var))

print("\n==== asizeof.asizeof() 결과 (내부 요소 포함) ====")
print("리스트:", asizeof.asizeof(list_var))
print("제너레이터:", asizeof.asizeof(gen_var))

print("\n==== 프로세스 전체 메모리 사용량 비교 ====")
show_memory("리스트 생성 후")
show_memory("제너레이터 생성 후")
````
