# 파이썬의 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소숫점이 없는 숫자)
* float: 실수(소숫점이 있는 숫자)
* str: 문자/문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음
* tuple: list와 비슷하나 한 번 만들면 수정이 안됨, 모든 자료형을 넣을 수 있음
* dictionary(dict): key : value 쌍으로 이루어진 자료형, 모든 자료형을 넣을 수 있음 
* set: 집합 자료형, 중복이 없는 자료를 만들 때, 모든 자료형을 넣을 수 있음, 중복 불가 {}안에 자료를 넣어서 만듬

# 리스트  list
* 리스트는 [정수, 실수, 문자, 리스트, 튜플 , 디셔너리 ] 모두 넣을 수있다.
* 리스트는 여러 자료가 같이 나열된 형태이므로 순서가 있다.
* 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 가능하다.

```python
l1 = [1,2.0, '삼', [1,2,3], (4,5,6), {"key":"value"}]
l1
```

```python
print(l1, type(l1))
```

# 리스트는 리스트 내의 자료형을 그대로 유지

```python
# 어떤 자료형(type)인지 수시로 확인해 봐야한다. 
for item in l1:
    print(item, type(item))
```

# 형변환 함수
* 형변환: data type 다른 data type으로 바꾸는 것

* 정수로 형변환 int()

```python
a = 3
b = "3"
print(a + b) # 형변환 함수를 이용해 일시적으로 정수로 변환
```

```python
a = 3
b = "3"
print(a + int(b)) # 형변환 함수를 이용해 일시적으로 정수로 변환
print("a: ", type(a), "b: ", type(b))
print("a: ", type(a), "int(b):", type(int(b)))
```

* 실수로 변환 float()

```python
print(b, type(b))
```

```python
# 3 + 3.0 = 6.0
a + float(b)
```

```python
c = 4.7
```

```python
a + c
```

```python
int(a + c)
```

```python
d = "3.4"
```

```python
print(d, type(d))
```

```python
# 실수 형태의 소숫점이 있는 문자는 float으로 변환해야 한다.
print(d, int(d)) 
```

* 문자열로 변환 str()
